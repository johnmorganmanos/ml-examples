```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
import numpy as np
import matplotlib.pyplot as plt

'''

This example comes from https://www.kaggle.com/carlmcbrideellis/lstm-time-series-prediction-sine-wave-example

Create a dataset of parts of a sine wave and train on it. How well does it predict the next step in the sequence?
Can we predict a phase shift?

'''
```

```python
X_train = np.arange(0,900,0.25) 
y_train = np.sin(X_train)

X_test = np.arange(600,1000,0.25) 
y_test = np.sin(X_test)

n_features = 1

train_series = y_train.reshape((len(y_train), n_features))
test_series  = y_test.reshape((len(y_test), n_features))
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 4))
ax.plot(X_train,y_train, lw=3, label='train data')
ax.plot(X_test, y_test,  lw=3, label='test data')
ax.legend(loc="lower left")
plt.show();
```

```python
look_back  = 100

train_generator = TimeseriesGenerator(train_series, train_series,
                                      length        = look_back, 
                                      sampling_rate = 1,
                                      stride        = 1,
                                      batch_size    = 10,
                                      shuffle       = False)

test_generator = TimeseriesGenerator(test_series, test_series,
                                      length        = look_back, 
                                      sampling_rate = 1,
                                      stride        = 1,
                                      batch_size    = 10,
                                      shuffle       = False)
```

```python
n_neurons  = 32

model = Sequential()
model.add(LSTM(n_neurons, input_shape=(look_back, n_features)))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')

model.fit(train_generator,epochs=50, shuffle = False)
```

```python
test_predictions  = model.predict(test_generator)
```

```python
x = np.arange(625,1000,0.25)
fig, ax = plt.subplots(1, 1, figsize=(15, 5))
ax.plot(X_train,y_train, lw=2, label='train data')
ax.plot(X_test,y_test, lw=3, c='y', label='test data')
ax.plot(x,test_predictions, lw=3, c='r',linestyle = ':', label='predictions')
ax.set_xlim([950, 1000])
ax.legend(loc="lower left")
plt.show();
```
